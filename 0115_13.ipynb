```python
from scipy.stats import chi2_contingency
```

```python
df_q2 = df.loc[df["major_discipline"] == "STEM", ].copy()
df_q2 = df_q2.loc[df_q2["city_development_index"] > df_q2["city_development_index"].quantile(0.85),]
```

```python
stat, p ,dof, exp_value = chi2_contingency(pd.crosstab(df_q2["relevent_experience"],
                                                       df_q2["target"]))
round(p,2)
```
